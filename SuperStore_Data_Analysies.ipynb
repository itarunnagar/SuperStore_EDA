#Starting Univariate Analysies

Convert Columns Names Order Date And Ship Date into Date Datatypes.

```python
# So first take the Order Date Columns And Extract the Years from Order Dates Columns ..
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])


# Create A new Columns With Names Order Year And Ship Years..
df['Order Year'] = df["Order Date"].dt.year
df['Ship Year'] = df["Ship Date"].dt.year
```

# Starting Bi-Variate Analysies


```python
# Numbers  of Order Per Category And Also Year Wise
ax = sns.countplot(x = 'Order Year' , data = df , hue = 'Category')

plt.title('Category Wise Order Per Year')
plt.xlabel('Years')
plt.ylabel('Numbers of Orders')

for unique in ax.containers :
  ax.bar_label(unique)

plt.savefig('Category_Wise_Order_Per_Year.jpg')
```

As We Know Year Wise All Three Categories will get more orders but Office Supplies Category will get more numbers of Orders...

```python
# Sales Per Years..
sales_per_years = df.groupby('Order Year')['Sales'].sum().reset_index()


plt.figure(figsize = (4, 4))
plt.bar(x = sales_per_years['Order Year'] , height = sales_per_years['Sales'] , data = sales_per_years , width = 0.5)
plt.title('Sales Per Years')
plt.xlabel('Years')
plt.ylabel('Sales')

plt.savefig('Sales_Per_Years.jpg')
plt.show()
```

Observation :- <br>
So we see our sales will increases year by years but in the 2014 The Sales was higher as compared to it Next Years...Else Sales Was Increases Year By Year

```python
# Profit Per Years..
per_per_year = df.groupby('Order Year')['Profit'].sum().reset_index()

# Make the Bar Plot...
plt.figure(figsize = (4, 4))
plt.bar(x = 'Order Year' , height = 'Profit' , data = per_per_year , width = 0.5)
plt.title('Profit Per Years')
plt.xlabel('Years')
plt.ylabel('Profit')

plt.savefig('Profit_Per_Years.jpg')
plt.show()
```

Observation :- <br>
Per Year Our Profit Increases Simultaneoutly


```python
# Total Stock Sales Per Years...
stock_sales_per_year = df.groupby('Order Year')['Quantity'].sum() .reset_index()

plt.figure(figsize = (4, 4))
plt.bar(x = stock_sales_per_year['Order Year'] , height = stock_sales_per_year['Quantity'] , data = stock_sales_per_year , width = 0.5)
plt.title('Stock Sales Per Years')
plt.xlabel('Years')
plt.ylabel('Stock Sales')

plt.savefig('Stock_Sales_Per_Years.jpg')
plt.show()
```

So we see the Stock Quantity was also increasing year by year that was so obivious
